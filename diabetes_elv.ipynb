    L1 = ...............

```python
def distance(i,p):
    L1 = Tn[Par].loc[i]
    L2 = Pn[Par].loc[p]
    diff = (L1-L2)**2
    return diff.sum()**0.5
```

<span style='color:white;font-weight:bold;background-color:#F0573B;padding: 2px;font-size:20px'>b.</span> Pour associer à chaque individu de <code>Tn</code> sa distance au patient <code>p</code>, et insérer la liste des résultats dans la  colonne <code>'d'</code> de <code>Tn</code>, exécuter le code suivant :

```python
Tn['d'] = [distance(i,0) for i in range(400)]
Tn.head()
```

<span style='color:white;font-weight:bold;background-color:#F0573B;padding: 2px;font-size:20px'>a.</span> Compléter les pointillés de la fonction suivante qui renvoie la prédiction concenrant le partient <code>p</code>.

def prediction(p):
    Tn['d'] = ........................................
    nb_malades = Tn.sort_values('d').head(11).M.sum()
    if nb_malades ..........:
        return 1
    else:

```python
def prediction(p):
    Tn['d'] = [distance(i,p) for i in range(400)]
    nb_malades = Tn.sort_values('d').head(11).M.sum()
    if nb_malades >= 6:
        return 1
    else:
        return 0
```

<h1 style="color:white;background-color:black;text-align:center"> Partie E : Différentes mesures de la fiabilité de prédiction </h1>

<span style='color:white;font-weight:bold;background-color:black;width:12px;padding: 2px;font-size:20px'>1.</span> Croisement des variables <code>'M'</code> et <code>'T'</code>
<br> Exécuter le code suivant pour obtenir la tabe croisée des effectifs des 2 variables <code>'M'</code> et <code>'T'</code>.

```python
E = pandas.crosstab(Pn['M'],Pn['T'],margins=True)
E
```

<span style='color:white;font-weight:bold;background-color:black;width:12px;padding: 2px;font-size:20px'>2.</span> Mesure de la pertinence du classement 

<span style='color:white;font-weight:bold;background-color:#F0573B;padding: 2px;font-size:20px'>a.</span> Calculer la précision du test, c'est à dire la probabilité d'être bien classé. 

```python
(E.loc[0,0]+E.loc[1,1])/100
```

<span style='color:white;font-weight:bold;background-color:#F0573B;padding: 2px;font-size:20px'>b.</span> Calculer la spécificité du test, c'est à est la probabilité d'avoir un test négatif sachant qu'on n'est pas malade

```python
E.loc[0,0]/50
```

<span style='color:white;font-weight:bold;background-color:#F0573B;padding: 2px;font-size:20px'>c.</span> Calculer la sensibilité du test c'est à dire la probabilité d'avoir un test positif sachant qu'on est malade

```python
E.loc[1,1]/50
```
